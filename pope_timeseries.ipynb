```python
import numpy as np
from datetime import datetime, date, timedelta
from random import uniform
import matplotlib.pyplot as plt
import pandas as pd
```

```python
date_ls = []
initial = datetime.strptime('2019-01-01', '%Y-%m-%d')
date_ls.append(initial)
for i in range(729):
    date_ls.append(date_ls[-1]+timedelta(days=1))
```

```python
# [print(date) for date in date_ls]
```

```python
rand_ls = []
for i in date_ls:
    rand_ls.append(int(uniform(-1000000,1000000)))
```

```python
plt.plot(date_ls,rand_ls)
plt.xticks(rotation = 90)
```

```python
plt.plot(date_ls[1:],np.diff(rand_ls,1)) # differences b/w adjacent days
plt.xticks(rotation = 90)
```

```python
plt.plot(date_ls[2:],np.diff(rand_ls,2)) # difference in difference b/w adjacent days
plt.xticks(rotation = 90)
```

```python
mths = pd.Series(range(24), index=pd.date_range("2019", freq="M", periods=24))
df = pd.DataFrame({"Time":date_ls,"Values":rand_ls})
```

```python
plt.plot(df.rolling(window = 30).mean()) # 30 day moving avg
```
